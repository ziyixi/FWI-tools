```python
lon_box=[91.3320117152011,74.6060844556399,174.409435753150,144.284491292185,91.3320117152011]
lat_box=[9.37366242174489,61.1396992149365,48.6744705245903,2.08633373396527,9.37366242174489]
coordinate=[]
for i,j in zip(lon_box,lat_box):
    phi=np.deg2rad(i)
    theta=np.deg2rad(90-j)
    x=np.sin(theta)*np.cos(phi)
    y=np.sin(theta)*np.sin(phi)
    z=np.cos(theta)
    coordinate.append((x,y,z))
    
sp=SphericalPolygon(coordinate)
def judge_if_in_box(lat,lon):
    phi=np.deg2rad(lon)
    theta=np.deg2rad(90-lat)
    x=np.sin(theta)*np.cos(phi)
    y=np.sin(theta)*np.sin(phi)
    z=np.cos(theta)
    return sp.contains_point((x,y,z))
```

```python
def generate_station_list():
    for network in network_log:
        if(len(network_log[network])!=0):
            client_name=""
            if("IRIS") in network_log[network]:
                client_name="IRIS"
            else:
                client_name=network_log[network][0]
                
            client=Client(client_name)
            inventory = client.get_stations(network=network, station="*",
                                        starttime=starttime,
                                        endtime=endtime,level="channel")
            contents=inventory.get_contents()["channels"]
            
            key_sets=set()
            for item in contents:
                network,station,_,_=item.split(".")
                if(".".join([network,station]) not in key_sets):
                    metadata=inventory.get_channel_metadata(item)
                    lat=metadata["latitude"]
                    lon=metadata["longitude"]
                    elv=metadata["elevation"]
                    dep=metadata["local_depth"]
                    
                    if(judge_if_in_box(lat,lon)):
                        print(station,network,round(lat,4),round(lon,4),elv,dep)
                key_sets.add(".".join([network,station]))
```
